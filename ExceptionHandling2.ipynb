```python
# Ans : 4
'''
LookupError class is is the base class for those exceptions that are
raised when a key or index used on a mapping or sequence is invalid or not found.
Exceptions :
            > KeyError
            > IndexError
'''
# Example for IndexError 

try :
    l=[1,2,3,4]
    print(l[4])
except IndexError as e :
    print(e)

# Example for KeyError

try:
    array={ 'a' : 1, 'b' : 2}
    print(array['c'])
except KeyError as e:
    print("Error occurs :",e)
```
